```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

# Define the range of n
n = np.arange(-10, 10)

# Define the midpoint for the unit impulse
mid = len(n) // 2

# Define h[n]
h_n = signal.unit_impulse(len(n), mid) - signal.unit_impulse(len(n), mid - 1)

# Define x[n]
x_n = np.heaviside(n, 1)

# Convolve x[n] and h[n] to get y[n]
y_n = np.convolve(x_n, h_n, mode='same')

# Plot h[n]
plt.figure(figsize=(10, 8))

plt.subplot(2, 1, 1)
plt.stem(n, np.real(h_n))
plt.title('Real Part of h[n]')
plt.grid(True)

plt.subplot(2, 1, 2)
plt.stem(n, np.imag(h_n))
plt.title('Imaginary Part of h[n]')
plt.grid(True)

plt.tight_layout()
plt.show()

# Plot x[n]
plt.figure(figsize=(10, 8))

plt.subplot(2, 1, 1)
plt.stem(n, np.real(x_n))
plt.title('Real Part of x[n]')
plt.grid(True)

plt.subplot(2, 1, 2)
plt.stem(n, np.imag(x_n))
plt.title('Imaginary Part of x[n]')
plt.grid(True)

plt.tight_layout()
plt.show()

# Plot y[n]
plt.figure(figsize=(10, 8))

plt.subplot(2, 1, 1)
plt.stem(n, np.real(y_n))
plt.title('Real Part of y[n]')
plt.grid(True)

plt.subplot(2, 1, 2)
plt.stem(n, np.imag(y_n))
plt.title('Imaginary Part of y[n]')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

# Define the range of n
n = np.arange(-10, 10)

# Define the midpoint for the unit impulse
mid = len(n) // 2

# Define h[n]
h_n = signal.unit_impulse(len(n), mid) - signal.unit_impulse(len(n), mid - 1)

# Define x[n]
x_n = np.where(n >= 0, n, 0)

# Convolve x[n] and h[n] to get y[n]
y_n = np.convolve(x_n, h_n, mode='same')

# Plot h[n]
plt.figure(figsize=(10, 8))

plt.subplot(2, 1, 1)
plt.stem(n, np.real(h_n))
plt.title('Real Part of h[n]')
plt.grid(True)

plt.subplot(2, 1, 2)
plt.stem(n, np.imag(h_n))
plt.title('Imaginary Part of h[n]')
plt.grid(True)

plt.tight_layout()
plt.show()

# Plot x[n]
plt.figure(figsize=(10, 8))

plt.subplot(2, 1, 1)
plt.stem(n, np.real(x_n))
plt.title('Real Part of x[n]')
plt.grid(True)

plt.subplot(2, 1, 2)
plt.stem(n, np.imag(x_n))
plt.title('Imaginary Part of x[n]')
plt.grid(True)

plt.tight_layout()
plt.show()

# Plot y[n]
plt.figure(figsize=(10, 8))

plt.subplot(2, 1, 1)
plt.stem(n, np.real(y_n))
plt.title('Real Part of y[n]')
plt.grid(True)

plt.subplot(2, 1, 2)
plt.stem(n, np.imag(y_n))
plt.title('Imaginary Part of y[n]')
plt.grid(True)

plt.tight_layout()
plt.show()
```
