# $\mathbb{Z}/n\mathbb{Z}$

Let $n$ be a positive integer.

$$
\mathbb{Z}/n\mathbb{Z} = \{k + n\mathbb{Z} : k \in \mathbb{Z}\} = \{ k + n \mathbb{Z}: 0 \leq k \leq n-1\}
$$

$\mathbb{Z}/n\mathbb{Z}$ is a ring. Let $(Z_n,+)$ be the additive group. $|Z_n|=n$.

# $\mathbb{C}^{Z_n}$, $L^1$, $L^2$ and Haar measure

Let

$$\mathbb{C}^{Z_n}$$

be the set of functions $Z_n \to \mathbb{C}$.

We use counting measure on the group $Z_n$ as Haar measure (which is discrete and compact) with total volume $n$, and  Since $Z_n$ has finite Haar measure (being a compact group) and every function $Z_n \to \mathbb{C}$ is continuous (being a discrete group), we have

$$\mathbb{C}^{Z_n}=L^1(Z_n)=L^2(Z_n).$$

We specify function space for emphasis.

For $f,g \in L^2(Z_n)$, define

$$
(f,g)_{L^2(Z_n)} = \sum_{x \in Z_n} f(x) \overline{g(x)} 
$$

Define

$$|f|_{L^2(Z_n)} = \sqrt{(f,f)_{L^2(Z_n)}}=\sqrt{\sum_{x \in Z_n} f(x) \overline{f(x)}}.$$

Define

$$
|f|_{L^1(Z_n)} = \sum_{x \in Z_n} |f(x)|.
$$

# $\delta_a$

For $a \in Z_n$, define $\delta_a:Z_n \to \mathbb{C}$ by

$$
\delta_a(x) = \begin{cases}
1&x=a\\
0&x \neq a
\end{cases}
$$

For $a,b \in Z_n$,

\begin{align*}
(\delta_a, \delta_b)_{L^2(Z_n)} &=\sum_{x \in Z_n} \delta_a(x) \overline{\delta_b(x)} \\
&=\sum_{x \in Z_n} \delta_a(x) \delta_b(x)\\
&= \begin{cases}
1&a = b\\
0&a \neq b
\end{cases}
\end{align*}

For $f \in L^2(Z_n)$,

$$
f(x) =\sum_{a \in Z_n} f(a) \delta_a(x) = \sum_{a=0}^{n-1} f(a) \delta_a(x),\qquad x \in Z_n.
$$

Indeed, $\{\delta_0,\ldots,\delta_{n-1}\}$ is an orthonormal basis of $L^2(Z_n)$.

# Convolution

For $f,g \in L^1(Z_n)$, define the **convolution** $f*g \in L^1(Z_n)$ by

$$
(f*g)(x) = \sum_{y \in Z_n} f(y)g(x-y),\qquad x \in Z_n.
$$

$$
f*g = g*f,\qquad f,g \in L^1(Z_n).
$$

$$
f*(g*h) = (f*g)*h,\qquad f,g,h \in L^1(Z_n).
$$

For $f \in L^1(Z_n)$ and for $a \in Z_n$,

$$
(f*\delta_a)(x) = f(x-a),\qquad x \in Z_n.
$$

For $a,b \in Z_n$, and for $x \in Z_n$,

$$
(\delta_a*\delta_b)(x)=\delta_a(x-b)=\delta_{a+b}(x).
$$

## Example

Take $n=15$. Define $f=\delta_0+\delta_1+\delta_2$. For $x \in Z_{15}$,

\begin{align*}
(f*f)(x)&=(\delta_0+\delta_1+\delta_2)*(\delta_0+\delta_1+\delta_2)\\
&=\delta_0 * \delta_0 + \delta_1*\delta_1 + \delta_2 * \delta_2\\
& + 2\delta_0 * \delta_1 + 2\delta_0 * \delta_2 + 2\delta_1*\delta_2\\
&=\delta_0 + \delta_2 + \delta_4\\
&+2\delta_1 + 2\delta_2 + 2 \delta_3\\
&=\delta_0 + 2\delta_1 + 3\delta_2 + 2\delta_3 + \delta_4
\end{align*}


# Dual group

Let $S^1=\{z \in \mathbb{C} : |z| = 1\}$, which is a multiplicative group.

Let $\widehat{Z_n}$ be the set of group homomorphisms $Z_n \to S^1$.

We use normalized counting measure on the group $\widehat{Z_n}$ as Haar measure (which is discrete and compact) with total volume 1.

For $a \in Z_n$, define $e_a:Z_n \to S^1$ by

$$
e_a(x) = \exp\left(2\pi i \dfrac{ax}{n}\right), \qquad x \in Z_n.
$$

$e_a$ is an element of $\widehat{Z_n}$.

Define $\psi:Z_n \to \widehat{Z_n}$ by $\psi(a)=e_a$. 

\begin{align*}
\widehat{Z_n} &= \{e_a : a \in Z_n\}\\
&= \{\psi(a): a \in Z_n\}\\
&= \psi(Z_n)
\end{align*}

$\psi:Z_n \to \widehat{Z_n}$ is an isomorphism of groups.

# Fourier transform

Define the **Fourier transform** $\mathscr{F}_n:L^2(Z_n) \to L^2(\widehat{Z_n)}$ by

$$
(\mathscr{F}_n f)(e_a) = \sum_{x \in Z_n} f(x) e_a(-x), \qquad e_a \in \widehat{Z_n}.
$$


# Pullback

We introduce the operator $F_n: L^2(Z_n) \to L^2(Z_n)$, which is defined via composition with the Fourier transform $\mathscr{F}_n$ and the function $\psi$ as

$$
F_n f = (\mathscr{F}_nf) \circ \psi.
$$

We pullback  $\mathscr{F}_nf:\widehat{Z_n} \to \mathbb{C}$ to a function $Z_n \to \mathbb{C}$.

We remind ourselves (a) that for $a \in Z_n$, the function $e_a:Z_n \to S^1$ is defined by

$$
e_a(x) = \exp\left(2\pi i \dfrac{ax}{n}\right), \qquad x \in Z_n,
$$

(b) that $\psi:Z_n \to \widehat{Z_n}$ is defined by $\psi(a)=e_a$,

and (c) that $\psi:Z_n \to \widehat{Z_n}$ is an isomorphism of groups, by

\begin{align*}
\widehat{Z_n} &= \{e_a : a \in Z_n\}\\
&= \{\psi(a): a \in Z_n\}\\
&= \psi(Z_n)
\end{align*}


\begin{align*}
(\mathscr{F}_nf)(\psi(a))&=\sum_{x \in Z_n} f(x) e_a(-x)\\
&=\sum_{x \in Z_n} f(x) \overline{e_a(x)}\\
&=(f,e_a)_{L^2(Z_n)}
\end{align*}

Thus

$$
(F_n f)(x) = (f,e_x)_{L^2(Z_n)}, \qquad x \in Z_n.
$$

In the sequel, we use the term Fourier transform to refer both to $\mathscr{F}_n$ and to $F_n$, but preserve the distinction for calculations.


## Example: $n=7$ and $f=\delta_3$

By

$$
(F_n f)(x) = (f,e_x)_{L^2(Z_n)}, \qquad x \in Z_n
$$

we have

$$
(F_7 \delta_3)(x) = (\delta_3,e_x)_{L^2(Z_7)}, \qquad x \in Z_7.
$$

The inner product $(\delta_3,e_x)_{L^2(Z_7)}$ is given by

\begin{align*}
(\delta_3,e_x)_{L^2(Z_7)} &= \sum_{y \in Z_7} \delta_3(y) \overline{e_x(y)}\\
&= \sum_{y=0}^{6} \delta_3(y) \overline{\exp\left(2\pi i \frac{xy}{7}\right)}.
\end{align*}

Since $\delta_3(y) = 1$ only when $y = 3$ and $0$ otherwise, the sum collapses to a single term:

\begin{align*}
(\delta_3,e_x)_{L^2(Z_7)} &= \overline{\exp\left(2\pi i \frac{3x}{7}\right)}\\
&= \exp\left(-2\pi i \frac{3x}{7}\right)\\
&= e_{-3}(x)
\end{align*}

Thus,

$$
(F_7 \delta_3)(x) = e_{-3}(x), \qquad x \in Z_7,
$$

namely,

$$
F_7 \delta_3 = e_{-3}.
$$

# signal.fftconvolve

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

n = 15

def cyclic_convolve(a, b, n):
    result = np.zeros(n, dtype=complex)
    for i in range(n):
        for j in range(n):
            result[i] += a[j] * b[(i - j) % n]
    return result

def plot_signals(title, f, g, h):
    n_domain = np.arange(n)
    fig, ax = plt.subplots(3, 2, figsize=(12, 8))

    ax[0,0].stem(n_domain, f.real, linefmt='black', markerfmt='ro', basefmt=' ')
    ax[0,0].annotate('f (Real)', (3, f.real[3]), textcoords="offset points", xytext=(0,10), ha='center')
    ax[0,0].axhline(0, color='black', linewidth=0.5)
    
    ax[0,1].stem(n_domain, f.imag, linefmt='black', markerfmt='ro', basefmt=' ')
    ax[0,1].annotate('f (Imag)', (3, f.imag[3]), textcoords="offset points", xytext=(0,10), ha='center')
    ax[0,1].axhline(0, color='black', linewidth=0.5)

    ax[1,0].stem(n_domain, g.real, linefmt='black', markerfmt='bo', basefmt=' ')
    ax[1,0].annotate('g (Real)', (6, g.real[6]), textcoords="offset points", xytext=(0,10), ha='center')
    ax[1,0].axhline(0, color='black', linewidth=0.5)
    
    ax[1,1].stem(n_domain, g.imag, linefmt='black', markerfmt='bo', basefmt=' ')
    ax[1,1].annotate('g (Imag)', (6, g.imag[6]), textcoords="offset points", xytext=(0,10), ha='center')
    ax[1,1].axhline(0, color='black', linewidth=0.5)

    ax[2,0].stem(n_domain, h.real, linefmt='black', markerfmt='ko', basefmt=' ')
    ax[2,0].annotate('f*g (Real)', (1, h.real[1]), textcoords="offset points", xytext=(0,10), ha='center')
    ax[2,0].axhline(0, color='black', linewidth=0.5)
    
    ax[2,1].stem(n_domain, h.imag, linefmt='black', markerfmt='ko', basefmt=' ')
    ax[2,1].annotate('f*g (Imag)', (1, h.imag[1]), textcoords="offset points", xytext=(0,10), ha='center')
    ax[2,1].axhline(0, color='black', linewidth=0.5)

    for a in ax[:,0]:
        a.set_ylabel('Amplitude', labelpad=10)
    for a in ax[2,:]:
        a.set_xlabel('Index')

    fig.suptitle(title)
    plt.tight_layout()
    plt.show()

# Define f and g
f = signal.unit_impulse(n, 3)
g = signal.unit_impulse(n, 6)

# Calculate the cyclic convolution of f and g
h = signal.fftconvolve(f, g, mode='same')

# Plotting the real and imaginary parts of f, g, and h
plot_signals('Signal Visualization', f, g, h)
```

# Fourier_transform

```python
import numpy as np

def Fourier_transform(n):
    """Create a Fourier transform function for the group Z_n."""

    def F(f):
        """Apply the Fourier transform to the function f."""

        def F_nf(x):
            """The Fourier transform of the function f at the point x."""
            return sum(f[j] * np.exp(-2j * np.pi * j * x / n) for j in range(n))

        # Calculate and return the Fourier-transformed function for all values in the domain
        return np.array([F_nf(x) for x in range(n)])

    # Return the Fourier transform function.
    return F
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

n = 7
n_domain = np.arange(0, n)

# Define f
f = signal.unit_impulse(n, 3)

# Calculate the Fourier transform of the signal
F = Fourier_transform(n)
transform = F(f)

# Plotting the real and imaginary parts
fig, ax = plt.subplots(2, 2, figsize=(10, 8))

# Plotting the real part of the signal
ax[0, 0].stem(n_domain, f.real)
ax[0, 0].set_xlabel('Index')
ax[0, 0].set_ylabel('Amplitude')
ax[0, 0].set_title('Real Part of the Signal')

# Plotting the imaginary part of the signal
ax[0, 1].stem(n_domain, f.imag)
ax[0, 1].set_xlabel('Index')
ax[0, 1].set_ylabel('Amplitude')
ax[0, 1].set_title('Imaginary Part of the Signal')

# Plotting the real part of the Fourier transform
ax[1, 0].stem(n_domain, transform.real)
ax[1, 0].set_xlabel('Frequency')
ax[1, 0].set_ylabel('Amplitude')
ax[1, 0].set_title('Real Part of the Fourier Transform')

# Plotting the imaginary part of the Fourier transform
ax[1, 1].stem(n_domain, transform.imag)
ax[1, 1].set_xlabel('Frequency')
ax[1, 1].set_ylabel('Amplitude')
ax[1, 1].set_title('Imaginary Part of the Fourier Transform')

plt.tight_layout()
plt.show()
```

# unitary_Fourier_transform

```python
import numpy as np

def unitary_Fourier_transform(n):
    """Create a unitary Fourier transform function for the group Z_n."""

    def Phi(f):
        """Apply the Fourier transform to the function f."""

        F = Fourier_transform(n)  # get the Fourier transform function
        F_f = F(f)  # compute the Fourier transform of f

        # normalize the Fourier-transformed function
        return F_f / np.sqrt(n)

    # Return the unitary Fourier transform function.
    return Phi
```

```python
import numpy as np
import scipy.signal as sp

n = 15

# Define f as a sum of delta functions
f = sp.unit_impulse(n, [0,1,2]) 

# Create Fourier_transform for n
F = unitary_Fourier_transform(n)

# Apply Fourier transform to f
f_transformed = np.array(F(f))

# Calculate the inner products
inner_product_space = np.vdot(f, f)
inner_product_freq = np.vdot(f_transformed, f_transformed)

print("Inner product in space domain: ", inner_product_space)
print("Inner product in frequency domain: ", inner_product_freq)
```

# Vandermonde matrix

```python
import numpy as np
from scipy import signal

n = 4
F_n = Fourier_transform(n)

# Compute the Fourier transform of unit impulses at each point in the domain
F_n_matrix = np.array([F_n(signal.unit_impulse(n, i)) for i in range(n)])

# Create a Vandermonde matrix and adjust it to match our Fourier matrix
x = np.exp(2j * np.pi * np.arange(n) / n)
V = np.vander(x, increasing=True)
V_adj = np.flip(V.conj(), axis=0)

print("Adjusted Vandermonde matrix:\n", V_adj)
print("\nFourier transform matrix:\n", F_n_matrix)

# Check if the determinant of the Fourier matrix 
# is close to the determinant of the adjusted Vandermonde matrix
print("\nAre the determinants close?", np.isclose(np.linalg.det(F_n_matrix), np.linalg.det(V_adj)))
```

```python
import numpy as np
from scipy import signal

n = 4
Phi_n = unitary_Fourier_transform(n)

# Compute the Unitary Fourier transform of unit impulses at each point in the domain
Phi_n_matrix = np.array([Phi_n(signal.unit_impulse(n, i)) for i in range(n)])

# Create a Vandermonde matrix, adjusted and scaled to match our Fourier matrix
x = np.exp(2j * np.pi * np.arange(n) / n)
V = np.vander(x, increasing=True) / np.sqrt(n)
V_adj = np.flip(V.conj(), axis=0)

print("Adjusted and scaled Vandermonde matrix:\n", V_adj)
print("\nUnitary Fourier transform matrix:\n", Phi_n_matrix)

# Check if the determinant of the Fourier matrix is close to the determinant of the adjusted Vandermonde matrix
print("\nAre the determinants close?", np.isclose(np.linalg.det(Phi_n_matrix), np.linalg.det(V_adj)))
```

# Quadratic Gauss sum

```python
import numpy as np
import matplotlib.pyplot as plt

def G(p):
    """Compute the Gauss sum G_p(x) as a function from Z_p to C."""
    return np.array([np.sum(np.fromiter((np.exp(2j * np.pi * x * (y**2) / p) for y in range(p)), dtype=complex)) for x in range(p)])

p = 13  # Change this value to compute for different primes
x_values = np.arange(p)

# Compute G_p and its Fourier transform
G_function = G(p)
F = Fourier_transform(p)
G_transform = F(G_function)

# Plotting the real and imaginary parts
fig, ax = plt.subplots(2, 3, figsize=(15, 10))

# Plotting the real part of G_p
ax[0, 0].stem(x_values, G_function.real)
ax[0, 0].set_xlabel('x')
ax[0, 0].set_ylabel('Amplitude')
ax[0, 0].set_title('Real Part of $G_p$')

# Plotting the imaginary part of G_p
ax[0, 1].stem(x_values, G_function.imag)
ax[0, 1].set_xlabel('x')
ax[0, 1].set_ylabel('Amplitude')
ax[0, 1].set_title('Imaginary Part of $G_p$')

# Plotting the absolute value of G_p
ax[0, 2].stem(x_values, np.abs(G_function))
ax[0, 2].set_xlabel('x')
ax[0, 2].set_ylabel('Magnitude')
ax[0, 2].set_title('Absolute Value of $G_p$')

# Plotting the real part of the Fourier transform
ax[1, 0].stem(x_values, G_transform.real)
ax[1, 0].set_xlabel('Frequency')
ax[1, 0].set_ylabel('Amplitude')
ax[1, 0].set_title('Real Part of the Fourier Transform of $G_p$')

# Plotting the imaginary part of the Fourier transform
ax[1, 1].stem(x_values, G_transform.imag)
ax[1, 1].set_xlabel('Frequency')
ax[1, 1].set_ylabel('Amplitude')
ax[1, 1].set_title('Imaginary Part of the Fourier Transform of $G_p$')

# Plotting the absolute value of the Fourier transform
ax[1, 2].stem(x_values, np.abs(G_transform))
ax[1, 2].set_xlabel('Frequency')
ax[1, 2].set_ylabel('Magnitude')
ax[1, 2].set_title('Absolute Value of the Fourier Transform of $G_p$')

plt.tight_layout()
plt.show()
```
