# PyWavelets

```python
import numpy as np
import matplotlib.pyplot as plt
import pywt

# Set the maximum level of wavelet coefficients to use in reconstruction
max_level = 1  # Adjust this value to suit your needs

# Create a sawtooth wave
x = np.linspace(0, 1, num=1024, endpoint=False)
y = np.where(x < 0.5, 2*x, 2*(1-x))

# Calculate wavelet coefficients
wavelet = pywt.Wavelet('haar')
coeffs = pywt.wavedec(y, wavelet)

# Zero out coefficients above the desired level
for i in range(max_level + 1, len(coeffs)):
    coeffs[i] = np.zeros_like(coeffs[i])

# Reconstruct the signal using the modified coefficients
y_reconstructed = pywt.waverec(coeffs, wavelet)

# Plot original and reconstructed signals
plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
plt.plot(x, y)
plt.title("Original Sawtooth Signal")

plt.subplot(2, 1, 2)
plt.plot(x, y_reconstructed)
plt.title(f"Reconstructed Signal (Up to Level {max_level})")

plt.tight_layout()
plt.show()
```
